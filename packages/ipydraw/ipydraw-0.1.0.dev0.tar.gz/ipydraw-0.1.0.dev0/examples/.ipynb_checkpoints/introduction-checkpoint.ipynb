```python
import base64
import io

import ipydraw
from PIL import Image
```

```python
w = ipydraw.Canvas(line_width=10, size=(500, 200), color=True)
w
```

```python
w.path
```

```python
syl = Image.open("IMG_2703.JPG")
syl.thumbnail((400, 400))
syl = syl.convert("RGBA")
syl
```

```python
mask = Image.open(io.BytesIO(base64.b64decode(w.data[len('data:image/png;base64,'):])))
mask.thumbnail((200, 200))
mask
```

```python
masked = syl.copy().convert("RGBA")
masked.alpha_composite(mask, (0, 0))
masked
```
